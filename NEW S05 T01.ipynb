```python
import numpy as np
import pandas as pd
import random
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
marathon_df = pd.read_csv('Dataset-Boston-2019.csv')
marathon_df
```

```python
marathon_df = marathon_df[['Age','Gender','Country','Result_hr']]
marathon_df
```

```python
marathon_df.describe().round()
```

```python
marathon_df['Gender'].value_counts()
```

```python
sns.set(style="darkgrid")
sns.set(rc= {'figure.figsize':(5,5)})
sns.histplot(data=marathon_df, x='Gender')
plt.show()
```

```python
plt.plot( 'x', 'y', "", data=marathon_df, linestyle='', marker='o')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
# muestra aleatoria simple
simple_marathon_df = marathon_df.sample(n = 500)
simple_marathon_df 
```

```python
simple_marathon_df.describe().round()
```

```python
simple_marathon_df['Gender'].value_counts()
```

```python
sns.set(style="darkgrid")
sns.set(rc= {'figure.figsize':(15,8)})
sns.histplot(data=simple_marathon_df, x='Gender')
plt.show()
```
